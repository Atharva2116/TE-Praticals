```python
import pandas as pd
```

```python

column=['Class','AGE','SEX','STEROID','ANTIVIRALS','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM','SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']
df=pd.read_csv("../25/hepatitis.csv",names=column)
df.head()
```

```python
import numpy as np
df.replace('?', np.nan, inplace=True)
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
```
