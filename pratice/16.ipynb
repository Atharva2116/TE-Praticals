```python
import pandas  as pd
import numpy as np
```

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
df = pd.read_csv("../16/adult.csv", header=None, names=column_names)
df.head()
```

```python
columns=['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week' ]
zscore_threshold = 3
```

```python
for column in columns:
    zscores=(df[column]-df[column].mean())/df[column].std()
    outliers=np.abs(zscores)>zscore_threshold
    df=df[~outliers]
    
df
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
cat_columns = ['workclass','education','marital-status','occupation','relationship','race','sex','native-country','income']

for column in cat_columns:
    le=LabelEncoder()
    df[column]=le.fit_transform(df[column])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
feature_columns=['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week' ]

target_column='income'
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[feature_columns],df[target_column],test_size=0.2,random_state=46)
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg_pred=reg.predict(x_test)
```

```python
print("Accuracy ",accuracy_score(y_test,reg_pred))
```

```python
nv=GaussianNB()
nv.fit(x_train,y_train)
```

```python
nv_pred=nv.predict(x_test)
```

```python
print("Accuracy ",accuracy_score(y_test,nv_pred))
```
