```python
import pandas as pd
import numpy as np
```

```python
columns=['Class','AGE','SEX','STEROID','ANTIVIRALS','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM','SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']
df=pd.read_csv("../26/hepatitis.csv",names=columns)
df.head()
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in df.columns:
    if df[column].dtype=="object":
        df[column]=le.fit_transform(df[column].astype(float))
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
# Remove rows with negative values
df=df[(df >=0 ).all(axis=1)]
```

Error correcting (Outlier detection and removal)

```python
mean_values=df.mean()
std_values=df.std()
```

```python
z_score=(df-mean_values)/std_values
```

```python
# Remove rows with Z-scores greater than 3 (outliers)
df=df[(np.abs(z_score)<3).all(axis=1)]
```

```python
df.head()
```

Data transformation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
df_scaled=scaler.fit_transform(df)
```

Build Data model using regression and Naïve Bayes methods for prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=df.drop('HISTOLOGY',axis=1)
y=df['HISTOLOGY']

```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=46)
```

```python
nv=GaussianNB()
nv.fit(x_train,y_train)
```

```python
nv_pred=nv.predict(x_test)
nv_pred
```

```python
lr=LogisticRegression()
lr.fit(x_test,y_test)
```

```python
lr_pred=lr.predict(x_test)
lr_pred
```

```python
print("Nabie bayes accuracy ",accuracy_score(y_test,nv_pred))
print("Logistic Regression accuracy ",accuracy_score(y_test,lr_pred))
```
