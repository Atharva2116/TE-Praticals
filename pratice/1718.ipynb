```python
import pandas as pd
import numpy as  np
```

```python
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
```

```python
df1=pd.read_csv("../17,18/Cleavland.csv",names=column_names)
df1.head()
```

```python
df2=pd.read_csv("../17,18/hung.csv" , names=column_names)
df2.head()
```

```python
df3=pd.read_csv("../17,18/Switzerland.csv", names=column_names)
df3.head()
```

```python
result=pd.concat([df1,df2,df3])
result.head()
```

```python
result.replace("?",np.nan,inplace=True)
```

```python
result.isna().sum()
```

```python
# result.fillna(0)
```

```python
result=result.astype(float)
```

```python
result.dtypes
```

```python
result['num']=result['num'].astype(int)
result['age']=result['age'].astype(int)
result['sex']=result['sex'].astype(int)
```

```python
result.dtypes
```

```python
# mean=result.mean()
# result.fillna(mean)
# result.fillna(0)
# null values
result['thalach'] = result['thalach'].fillna(0)
result['chol'] = result['chol'].fillna(0)
result['fbs'] = result['fbs'].fillna(0)
result['restecg'] = result['restecg'].fillna(0)
result['trestbps'] = result['trestbps'].fillna(0)
result['exang'] = result['exang'].fillna(0)
result['oldpeak'] = result['oldpeak'].fillna(0)
result['slope'] = result['slope'].fillna(0)
result['ca'] = result['ca'].fillna(0)
result['thal'] = result['thal'].fillna(0)
result.isnull().sum()


```

```python
result.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

HISTOGRAMS


```python
sns.histplot(x=result['age'],data=result,multiple="stack",hue="cp")
```

 Bar Plots 
 Objective: Compare the values of a categorical variable, such as the frequency of different types of heart disease, or the distribution of sexes.

```python
plt.figure(figsize=(8, 6))
result['num'].value_counts().plot(kind='bar', color='lightgreen')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
result['sex'].value_counts().plot(kind='bar', color='pink')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Sex Distribution')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=result,x="age",y="chol",palette="pastel",errorbar=None)
plt.show()
```

Line Charts  (Visualize trends or patterns in a continuous variable, such as heart rate (thalach) over time.)

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=result,x="age",y="thalach",color='purple')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=result,x="age",y="chol",color='purple')
plt.show()
```

Add Histogram and Scatter plot to box plot. 

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Create box plot
sns.boxplot(x='sex', y='age', data=result, ax=ax)

# Add histogram on top of the box plot
sns.histplot(data=result, x='age', hue='sex', bins=20, kde=True, alpha=0.5, ax=ax)

# Add scatter plot at the bottom
sns.scatterplot(data=result, x='age', y='chol', hue='sex', alpha=0.5, ax=ax)

# Set titles and labels
ax.set_title('Box plot with Histogram and Scatter plot')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')

# Show the plot
plt.show()
```

Pie chart
