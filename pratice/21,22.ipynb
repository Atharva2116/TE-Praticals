```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
df=pd.read_csv("../21-22/adult.csv",names=column_names)
df.head()
```

 Histograms 

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df,x=df["income"],hue="occupation",multiple="stack")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# plt.hist(df['age'],bins=20,color="lightblue",edgecolor="black")
plt.hist(x=df['income'],bins=20,color="lightblue",edgecolor="black")
plt.title('Distribution of Age')
plt.xlabel("AGE")
plt.ylabel("Frequncy")
plt.show()
```

Dot Plots


```python
plt.figure(figsize=(18, 10))
sns.stripplot(data=df,x="education",y="education-num",jitter=True,palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['education-num'],df['education'],'bo',alpha=0.5)
plt.title('Education Level Dot Plot')
plt.xlabel('Education Number')
plt.ylabel('Education Level')
plt.show()
```

Bar plot

```python
plt.figure(figsize=(18,6))
sns.barplot(data=df,x="education",y="education-num",palette="pastel")
```

Line charts

```python
plt.figure(figsize=(18,4))
sns.lineplot(data=df,x="education",y="hours-per-week",color="purple")
plt.show()
```

Add Histogram and Scatter plot to box plot.


Pie Charts

```python
mlab = ['Married-civ-spouse', 'Divorced','Never-married','Separated','Widowed','Married-spouse-absent','Married-AF-spouse']
plt.figure(figsize=(8,8))
df['marital-status'].value_counts().plot(kind="pie",autopct="%1.1f%%",labels=mlab)
plt.show()
```

Box plot

```python
plt.figure(figsize=(25,8))
sns.boxplot(data=df,x='hours-per-week',y='education-num',palette="pastel")
plt.plot()
```

Scatter Plot

```python
sns.scatterplot(data=df,x='education-num',y='capital-loss',palette="pasetel")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='education-num',y='education',data=df)
plt.title('Education Level vs Hours per Week')
plt.xlabel('Education Number')
plt.ylabel('Hours per Week')
plt.show()
```
